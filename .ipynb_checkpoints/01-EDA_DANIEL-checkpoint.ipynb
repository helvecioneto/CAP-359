#### Bibliotecas

```python
import os
import gdal
import osr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
from sklearn.cluster import DBSCAN
from shapely.geometry import Polygon, MultiPoint, MultiLineString
from shapely.geometry import Point
```

#### 1 - Exemplo abrir e plotar dados

```python
file = "data/parcela/tif/vb_pac1_nasa.tif"
ds = gdal.Open(file)
band = ds.GetRasterBand(1)
arr = band.ReadAsArray()
arr[arr == -9999] = np.nan
```

```python
# unravel GDAL affine transform parameters
c, a, b, f, d, e = ds.GetGeoTransform()

def pixel2coord(col, row):
    xp = a * col + b * row + a * 0.5 + b * 0.5 + c
    yp = d * col + e * row + d * 0.5 + e * 0.5 + f
    return(xp, yp)
```

```python
plt.imshow(arr)
```

#### Clustering

```python
### Threshold limiarization
def pre_processing(data,threshold):
    np.warnings.filterwarnings('ignore')
    
    rs =  np.where(data >= threshold)
    rs = np.asanyarray(rs)
    pe = pd.DataFrame({'X':rs[1],'Y':rs[0]})
    
    alt = []
    
    for i,row in pe.iterrows():
        alt.append(data[row.Y][row.X])
    
    pe['ALT'] = alt
    
    return pe
```

```python
def clusterization(time):
    te = time.copy()
    
    te = time[['X','Y']].values
    output = pd.DataFrame()

    ms = DBSCAN(algorithm='auto', eps=2, metric='euclidean',
    metric_params=None, min_samples=5, n_jobs=None, p=None)
    
    ms.fit(te)
    labels = ms.labels_
    n_clusters_ = len(np.unique(labels))
    output['CLUSTER']=labels
    output['X'] = te[:,0]
    output['Y'] = te[:,1]
    output['ALT'] = time['ALT']
    
    output = output[output.CLUSTER != -1]

    return output
```

```python
pre_data = pre_processing(arr,12)
clu_data = clusterization(pre_data)
# Remove cluster less than 3 elements
clu_data = clu_data.groupby('CLUSTER').filter(lambda x : len(x)>3)
```

```python
clu_data.head()
```

```python
from shapely.geometry import Polygon, MultiPoint, MultiLineString, MultiPolygon
from shapely.geometry import Point
import geopandas as gpd
import alphashape
import shapely
```

```python
clu_data['geometry'] = clu_data.apply(lambda x: pixel2coord(x.X, x.Y), axis=1)
clu_data.geometry = clu_data.geometry.apply(Point)

pointdata = gpd.GeoDataFrame(clu_data, geometry='geometry')
pointdata.crs = {'init' :'epsg:32721'}
```

```python
clu_data['geom'] = list(zip(clu_data.X,clu_data.Y))
clu_data.geom = clu_data.geom.apply(Point)
clu_data.head()
```

```python
## Get values from concave_hull
lc = clu_data.CLUSTER.unique()
ll = []

for i in lc:
    ll.append(alphashape.alphashape(clu_data.loc[clu_data['CLUSTER'] == i]['geom']))
```

```python
gp = pd.DataFrame(columns=['CLUSTER','LAT','LON'])

for i in range(len(ll)):
    x = ll[i].exterior.xy[0],ll[i].exterior.xy[1]
    for j in range(len(x[0])):
        aa = (pixel2coord(x[0][j],x[1][j]))
        gp = gp.append({'CLUSTER': i,'LAT':aa[0],'LON':aa[1]}, ignore_index=True)
```

```python
gp
```

```python
gp['geom'] = list(zip(gp.LAT,gp.LON))

gpD = pd.DataFrame()
# gp['geometry'] = gp['geometry'].apply(lambda x: x.coords[0])
gpD['geometry'] = gp.groupby('CLUSTER')['geom'].apply(lambda x: Polygon(x.tolist()))
cpdata = gpd.GeoDataFrame(gpD, geometry = 'geometry')
cpdata.crs = {'init' :'epsg:32721'}
cpdata['AREA'] = cpdata.area
cpdata = cpdata.reset_index()
cpdata['CLUSTER'] = cpdata.CLUSTER.astype(int)
```
