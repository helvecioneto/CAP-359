```python
import numpy as np
import matplotlib.pyplot as plt
import animatplot as amp

%matplotlib nbagg
```

```python
## Read data times
t1 = np.genfromtxt('data/t1.csv', delimiter=',')
t2 = np.genfromtxt('data/t2.csv', delimiter=',')
t3 = np.genfromtxt('data/t3.csv', delimiter=',')
t4 = np.genfromtxt('data/t4.csv', delimiter=',')
t5 = np.genfromtxt('data/t5.csv', delimiter=',')

lat = np.genfromtxt('data/lat.csv', delimiter=',')
lon = np.genfromtxt('data/lon.csv', delimiter=',')

## Values 0 to np.nan
t1[t1 == 0] = np.nan
t2[t2 == 0] = np.nan
t3[t3 == 0] = np.nan
t4[t4 == 0] = np.nan
t5[t5 == 0] = np.nan
```

```python
frames = [t1,t2,t3,t4,t5]
type(frames)
```

```python
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
t = np.linspace(0, 2*np.pi, 40)

X, Y, T = np.meshgrid(x, y, t)

Z = np.sin(X*X+Y*Y-T)

block = amp.blocks.Pcolormesh(X[:,:,0], Y[:,:,0], Z, t_axis=2, cmap='RdBu')
plt.colorbar(block.quad)
plt.gca().set_aspect('equal')

anim = amp.Animation([block], amp.Timeline(t))

anim.controls()
```
